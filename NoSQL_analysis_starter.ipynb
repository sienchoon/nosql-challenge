```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name

showdb = mongo.list_database_names()
print(showdb)
```

```python
# review the collections in our database
ukfood = mongo['uk_food']
print(ukfood.list_collection_names())
```

```python
# assign the collection to a variable
establishments = ukfood['establishments']
establishments.find_one()
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}

# Use count_documents to display the number of documents in the result
doc_no = establishments.count_documents(query)
print(f'Number of documents: {doc_no}')
# Display the first document in the results using pprint
results = list(establishments.find(query))
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
rows = results_df.shape[0]
print(f'Total number of rows in DataFrame: {rows}')
# Display the first 10 rows of the DataFrame
results_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': 'London'}, 'RatingValue': {'$gte': 4}}

# Use count_documents to display the number of documents in the result
doc_no = establishments.count_documents(query)
print(f'Number of documents with London as Local Authority and RatingValue >= 4: {doc_no}') 
# Display the first document in the results using pprint
results = list(establishments.find(query))
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
london_gte4_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
rows = london_gte4_df.shape[0]
print("Number of rows in DataFrame: ", rows)
# Display the first 10 rows of the DataFrame
london_gte4_df.head(10)
```
