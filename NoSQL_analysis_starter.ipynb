### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}

# Use count_documents to display the number of documents in the result
doc_no = establishments.count_documents(query)
print(f'Number of documents: {doc_no}')
# Display the first document in the results using pprint
results = list(establishments.find(query))
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
rows = results_df.shape[0]
print(f'Total number of rows in DataFrame: {rows}')
# Display the first 10 rows of the DataFrame
results_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': 'London'}, 'RatingValue': {'$gte': 4}}

# Use count_documents to display the number of documents in the result
doc_no = establishments.count_documents(query)
print(f'Number of documents with London as Local Authority and RatingValue >= 4: {doc_no}') 
# Display the first document in the results using pprint
results = list(establishments.find(query))
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
london_gte4_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
rows = london_gte4_df.shape[0]
print("Number of rows in DataFrame: ", rows)
# Display the first 10 rows of the DataFrame
london_gte4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
query = {'BusinessName' : "Penang Flavours"}
penang_flavour = list(establishments.find(query))
pprint(penang_flavour)
```

```python
# Remove duplicates from the collection
find = establishments.count_documents({"BusinessName": "Penang Flavours"})
print(f'There are a total of {find} duplicate Penang Flavours in the UK food database') # 1 (find)
```

```python
business_name = 'Penang Flavours'
pipeline = [
    {
        '$match': {
            'BusinessName': business_name  # Match documents with the specified BusinessName
        }
    },
    {
        '$group': {
            '_id': '$BusinessName',  # Group by BusinessName
            'documentId': {'$first': '$_id'},  # Keep the ObjectId of the first document in each group
            'count': {'$sum': 1}  # Count the number of documents in each group
        }
    },
    {
        '$match': {
            'count': {'$gt': 1}  # Filter groups with more than one document (duplicates)
        }
    },
    {
        '$project': {
            '_id': '$documentId'  # Project the ObjectId of the first document as the final result
        }
    }
]

results = establishments.aggregate(pipeline)
duplicate_ids = [result['_id'] for result in results]

delete_duplicate = establishments.delete_many({'_id': {'$in': duplicate_ids}})

print(f'Removed {delete_duplicate.deleted_count} duplicate for {business_name}')

```

```python
pprint(penang_flavour)
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

greater_latitude = latitude + degree_search
lesser_latitude = latitude - degree_search

greater_longitude = longitude + degree_search
lesser_longitude = longitude - degree_search

# Query latitude = greater than 51.49014200 - 0.01 but less than 51.49014200 + 0.01
# Query longitude = greater than 0.08384000 - 0.01 but less than 0.08384000 + 0.01
query = {"RatingValue": {"$eq" : 5}, 
         "geocode.latitude": {"$gte": lesser_latitude, "$lte": greater_latitude}, 
         "geocode.longitude": {"$gte": lesser_longitude, "$lte": greater_longitude}}
sort =  [("scores.Hygiene", 1)]

top_5 = 5
# Print the results
results = list(establishments.find(query).sort(sort).limit(top_5))
print(f'There are {len(results)} establishments with a Rating Value of 5 within 0.01 degree of {latitude}, {longitude}')

```

```python
# Convert result to Pandas DataFrame
top_5_establishments_df = pd.DataFrame(results)
top_5_establishments_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
hygiene_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}
# 2. Groups the matches by Local Authority
group_LA = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}
# 3. Sorts the matches from highest to lowest
sort_LA = {'$sort': {'count': -1}}
# Print the number of documents in the result
results = list(establishments.aggregate([hygiene_query, group_LA, sort_LA]))
print(f'Number of documents in result: {len(results)}')
# Print the first 10 results
pprint(results[:10])
```

```python
# Convert the result to a Pandas DataFrame
group_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
rows = group_df.shape[0]
print(f'There are {rows} rows in the DataFrame')
# Display the first 10 rows of the DataFrame
group_df.head(10)
```
